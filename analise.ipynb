```python
import pandas as p
tabel=p.read_csv("terrorist.csv")
import matplotlib.pyplot as plt

```

```python
#removing unwanted columns
#removendo colunas indesejadas
tabel=tabel.drop("Code",axis=1)
display(tabel)
```

```python
#total attacks in the table
#total de ataques na tabela
total=tabel['Terrorist attacks'].sum()
text = " Total Terrorist attacks"
plt.figure(figsize=(2,3))  
plt.text(0.5, 0.6, text, fontsize=16, ha='center')  
plt.text(0.5, 0.4, (total), fontsize=30, ha='center')  
plt.axis('off')  
plt.show()
```

```python

# Calcular a quantidade total de ataques terroristas por paÃ­s
# Calculate the total amount of terrorist attacks by country
totalpais = tabel.groupby('Entity')['Terrorist attacks'].sum()
for pais, total_ataques in totalpais.items():
    print(f'country,continents: {pais} - Total Atacks: {total_ataques}')
    
#Creating graph
#criando grafico
totalcountry = tabel.groupby('Entity')['Terrorist attacks'].sum()
print(totalcountry)
plt.figure(figsize=(75, 10))
totalcountry.plot(kind='bar')
plt.title('number of incidents by continents and countries')
plt.xlabel('Country')
plt.ylabel('Total  incidents')
plt.show()
```

```python
#average number of attacks per country in the table
#mÃ©dia de ataques por paÃ­s na tabela
paises = tabel['Entity'].tolist()
continents = ['Africa', 'Asia'] 
countr = [pais for pais in paises if pais not in continents]
paisesunicos = set(countr)
quantidadepaises = len(paisesunicos)
med=total/quantidadepaises
print(med)

printmed=med
text2 = " average number of attacks per country in the table"
plt.figure(figsize=(1.5,3))  
plt.text(0.1, 0.6, text2, fontsize=11, ha='center')  
plt.text(0.5, 0.4, (printmed), fontsize=30, ha='center')  
plt.axis('off')  
plt.show()

text3 ="Total of countries (approximately)"
plt.figure(figsize=(1,1))  
plt.text(0.1, 0.9, text3, fontsize=11, ha='center')  
plt.text(0.5, 0.4, (quantidadepaises), fontsize=30, ha='center')  
plt.axis('off')  
plt.show()
```

```python
    
# Calcular a quantidade total de ataques terroristas por Ano
# Calculate the total amount of terrorist attacks by Year
totalcountry = tabel.groupby('Year')['Terrorist attacks'].sum()
print(totalcountry)
plt.figure(figsize=(75, 10))
totalcountry.plot(kind='bar')
plt.title('number of incidents by Year')
plt.xlabel('Year')
plt.ylabel('Total attacks')
plt.show()
```
